```python
'''
The relationship between gravitational force and distance between two bodies
'''
import matplotlib.pyplot as plt
def draw_graph(x,y):
  plt.plot(x,y,marker='o')
  plt.xlabel('Distance in meters')
  plt.ylabel('Gravitational force in newtons')
  plt.title('Gravitational force and distance')
  plt.show()
  
def generate_F_r():
  r=range(100,1001,50)#100以上1001未満の中で50ずつ増やすリスト
  F=[]
  G=6.674*(10**-11)
  m1=0.5
  m2=1.5
  for dist in r:
    force=G*(m1*m2)/(dist**2)
    F.append(force)#空リストFに対してforceを加えていく
  draw_graph(r,F)
```

```python
generate_F_r()
```

```python
'''
ここはグラフを書く準備
'''
from matplotlib import pyplot as plt
import math
def draw_graph2(x,y):
  plt.plot(x,y,marker='o')# 【演習】marker=''を付けると，グラフが変化します
  plt.xlabel('x-coordinate')
  plt.ylabel('y-coordinate')
  plt.title('Projectitle motion of a ball')

def frange(start,final,interval):
  numbers=[]
  while start<final:
    numbers.append(start)
    start=start+interval
  return numbers
'''
Remark
range(n,m,k)は整数値にしか対応していないため，新しくfrange()を作ってrangeと同じ働きをするようにプログラムした

ここから投射の軌跡の出力
'''

def draw_trajectory(u,theta):
  theta=math.radians(theta)
  g=9.8
  t_flight=2*u*math.sin(theta)/g
  intervals=frange(0,t_flight,0.001)
  x=[]
  y=[]
  for t in intervals:
    x.append(u*math.cos(theta)*t)
    y.append(u*math.sin(theta)*t-0.5*g*t*t) # あくまで二次関数をプロットしているわけではなく，パラメータ表示のリストをプロットしているだけ．
  draw_graph2(x,y)
if __name__=='__main__':#うまくプログラムが動くようにするおまじない
  try:
    u=float(input('初期速度を入力してください：'))
    theta=float(input('投射角を入力してください：'))
  except ValueError:
    print('その値は入力不可能です')
  else:
    draw_trajectory(u,theta)
    plt.show()
```

```python
u_list=[30,40,60]
theta=45
for u in u_list:
  draw_trajectory(u,theta)

plt.legend(['30','40','60'])
plt.show()
```

```python
def draw_graph3(x,y):
  plt.plot(x,y)
  plt.xlabel('x')
  plt.ylabel('y')
  plt.title('plot circle')
  plt.axes().set_aspect('equal')# アスペクト比をそろえる(これないと，楕円に見えちゃう)

def circle(p,q,r): # (x-p)^2+(y-q)^2=r^2
  x=[]
  y=[]
  intervals=frange(0,2*3.14,0.01)
  for i in intervals:
    x.append(r*math.cos(i)+p)
    y.append(r*math.sin(i)+q)# 円のパラメータ表示
  draw_graph3(x,y)
if __name__=='__main__':
  try:
    p=float(input('中心x座標'))
    q=float(input('中心y座標'))
    r=float(input('半径'))
  except ValueError:
    print('だめ')
  else:
    circle(p,q,r)
    plt.show()
```

```python
plist=[1,2,1,0]
qlist=[1,0,-1,0]
for i in range(0,4):
  circle(plist[i],qlist[i],1)
```
